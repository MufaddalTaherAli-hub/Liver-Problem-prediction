## Preprocess

```python
df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_gender = LabelEncoder()

df['gender'] = le_gender.fit_transform(df['gender'])
```

```python
df.head(2)
```

# machine learning

```python
x = df.drop('is_patient', axis=1)
y= df['is_patient']
```

```python
print(x.shape)
print(y.shape)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE()
x_res, y_res = sm.fit_resample(x, y)
```

```python
y_res.value_counts()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_res, y_res, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

__1. Logistic Regression__

```python
model_log = LogisticRegression()
model_log.fit(X_train, y_train)
```

```python
log_y_pred = model_log.predict(X_test)
```

```python
from sklearn import metrics
```

```python
log_cm = metrics.confusion_matrix(y_test, log_y_pred)
print(log_cm)
print('-'*100)
sns.heatmap(log_cm, annot=True, fmt='.0f', color = 'b')
plt.show()
```

```python
print(metrics.classification_report(y_test, log_y_pred))
```

```python
log_score = metrics.accuracy_score(y_test, log_y_pred)
log_score
```

__2. Decision Tree__

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_tree = DecisionTreeClassifier()
model_tree.fit(X_train, y_train)
```

```python
tree_y_pred = model_tree.predict(X_test)
```

```python
tree_cm = metrics.confusion_matrix(y_test, tree_y_pred)
print(tree_cm)
print('-'*100)
sns.heatmap(tree_cm, annot=True, fmt='.0f', color = 'b')
plt.show()
```

```python
print(metrics.classification_report(y_test, tree_y_pred))
```

```python
tree_score = metrics.accuracy_score(y_test, tree_y_pred)
tree_score
```

__3. Random Forest__

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_forest = RandomForestClassifier()
model_forest.fit(X_train, y_train)
```

```python
forest_y_pred = model_forest.predict(X_test)
```

```python
forest_cm = metrics.confusion_matrix(y_test, forest_y_pred)
print(forest_cm)
print('-'*100)
sns.heatmap(forest_cm, annot=True, fmt='.0f', color = 'b')
plt.show()
```

```python
print(metrics.classification_report(y_test, forest_y_pred))
```

```python
forest_score = metrics.accuracy_score(y_test, forest_y_pred)
forest_score
```

__4. KNeighborsClassifier__

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_neighbor = KNeighborsClassifier()
model_neighbor.fit(X_train, y_train)
```

```python
knn_y_pred = model_neighbor.predict(X_test)
```

```python
knn_cm = metrics.confusion_matrix(y_test, knn_y_pred)
print(knn_cm)
print('-'*100)
sns.heatmap(knn_cm, annot=True, fmt='.0f', color = 'b')
plt.show()
```

```python
print(metrics.classification_report(y_test, knn_y_pred))
```

```python
knn_score = metrics.accuracy_score(y_test, knn_y_pred)
knn_score
```

__Random Forest has a highest accuracy__
